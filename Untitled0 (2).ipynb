```python
import torch
import torch.nn as nn
import torch.optim as optim
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(3, 4)
        self.fc2 = nn.Linear(4, 4)
        self.fc3 = nn.Linear(4, 3)
        self.out = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()

    def forward(self, x):
        x = self.sigmoid(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        x = self.tanh(self.out(x))
        return x

model = SimpleNN()
input_data = torch.rand((100, 3))
predictions_before_training = model(input_data).detach().numpy()

sns.lineplot(x=np.arange(100), y=predictions_before_training.flatten())
plt.title('Predictions Before Training')
plt.xlabel('Sample')
plt.ylabel('Prediction')
plt.show()

for param in model.parameters():
    param.requires_grad = True

loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
target_output = torch.rand((100, 1))
initial_loss = loss_fn(model(input_data), target_output)
print(f'Initial Loss: {initial_loss.item()}')

sns.lineplot(x=np.arange(100), y=predictions_before_training.flatten())
plt.title('Initial Predictions Before Training')
plt.show()

epochs = 100
losses = []

for epoch in range(epochs):
    optimizer.zero_grad()
    predictions = model(input_data)
    loss = loss_fn(predictions, target_output)
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')
    losses.append(loss.item())

plt.plot(range(epochs), losses)
plt.title('Loss Over Time')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

predictions_after_training = model(input_data).detach().numpy()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.lineplot(x=np.arange(100), y=predictions_before_training.flatten())
plt.title('Predictions Before Training')

plt.subplot(1, 2, 2)
sns.lineplot(x=np.arange(100), y=predictions_after_training.flatten())
plt.title('Predictions After Training')

plt.show()
```
